```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Times New Roman
matplotlib.rc('font', family='Times New Roman')

# Load the CSV file
file_path = './building.csv'
data = pd.read_csv(file_path)


# 
plt.figure(figsize=(3, 3),dpi=300)
plt.scatter(data['building_google'], data['building_baidu'], alpha=0.5, label='building',s=0.01)
plt.title('building scatter')
plt.xlabel('ID')
plt.ylabel('Google building')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True)


# r2
r2 = r2_score(data['id'], data['building_google'])

plt.show(), r2
```
