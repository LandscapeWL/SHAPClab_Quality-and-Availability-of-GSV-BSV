```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 设置字体为Times New Roman
matplotlib.rc('font', family='Times New Roman')
```

# building

```python
# Load the CSV file
data = pd.read_csv('./building.csv')

# 提取需要计算相关性的两列
google = data['building_google']
baidu = data['building_baidu']

# 计算皮尔逊相关系数和p-value
r_value, p_value = pearsonr(google, baidu)

# 打印结果
print('皮尔逊相关系数:', r_value)
print('p-value:', p_value)


plt.figure(figsize=(3, 3),dpi=300) # 设置画布

plt.scatter(google, baidu,
            alpha=1,
            s=0.1,
            zorder=2,
            edgecolors='none',
            facecolors='#da0018')

plt.title('Building Index')
plt.xlabel('GSV')
plt.ylabel('BSV')

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True, zorder=1) # 绘制网格

plt.text(0.6, 0.9, f'R = {r_value:.3f}', fontsize=16, color='#da0018') # 绘制r

plt.show()
```
