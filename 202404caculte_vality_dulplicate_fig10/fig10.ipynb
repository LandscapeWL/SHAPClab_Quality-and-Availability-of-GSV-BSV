# 绘制有效率的对比折线图

```python
import matplotlib.pyplot as plt

# 数据
intervals = ['5m', '10m', '20m', '30m', '40m', '50m', '100m', '200m']
google_valid = [52.3483, 52.6223, 53.2178, 53.8435, 54.4598, 54.9605, 56.8886, 58.7398]
baidu_valid = [57.6115, 58.091, 59.1596, 60.2288, 61.297, 62.1528, 65.5212, 68.7936]

# 创建图表
plt.figure(figsize=(4.5, 4),dpi=300)
plt.plot(intervals, google_valid, label='Google Valid (%)', marker='o', color='#21a462') # 使用指定颜色
plt.plot(intervals, baidu_valid, label='Baidu Valid (%)', marker='o', color='#f72c30') # 使用指定颜色

# 添加标题和标签
plt.title('Validity Comparison Over Different Intervals')
plt.xlabel('Interval')
plt.ylabel('Validity Percentage (%)')

# 添加四周边框
plt.gca().spines['top'].set_visible(True)
plt.gca().spines['right'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(True)

# 去掉背景辅助线
plt.grid(False)

# 显示图例
plt.legend()

# 展示图表
plt.show()
```
